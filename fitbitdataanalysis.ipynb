checking the distribution of numerical feature

```python
numeric = [feature for feature in df.columns if df[feature].values.dtype !='O']
numeric
```

```python
plt.figure(figsize=(20,20))
for n, feature in enumerate(numeric):
    if feature != 'Id':
        ax = plt.subplot(5,3,n)
        sns.histplot(ax=ax,bins=50,data=df, x =feature)

```

### In above histograms we can see that some of the features look normal and some are not normal. We can also see that LoggedActivites Distance is 0 for almost all data so we can drop that

```python
# lets check for the outliers in the dataset
plt.figure(figsize=(15,10))
for n,feature in enumerate(numeric):
    if feature  != 'Id':
        ax = plt.subplot(5,3,n)
        sns.boxplot(x = feature, data = df ,ax=ax)
        plt.subplots_adjust(left=0.1,
                    bottom=0.0, 
                    top = 0.8
                   )
plt.show()
        
```

## It looks like there are some outliers in the data which can be deleted to improve any machine learning  model
